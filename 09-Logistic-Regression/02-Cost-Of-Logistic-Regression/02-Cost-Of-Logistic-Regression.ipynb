## 逻辑回归的损失函数

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def costy1(x):
    return -np.log(x)
def costy0(x):
    return -np.log(1-x)
```

```python
x = np.linspace(0, 1, 50)

plt.plot(x, costy1(x), label="y=1")
plt.plot(x, costy0(x), label="y=0")
plt.legend()
plt.show()
```
