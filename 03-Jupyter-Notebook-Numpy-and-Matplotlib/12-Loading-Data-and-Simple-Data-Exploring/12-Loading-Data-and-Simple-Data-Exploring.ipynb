## 12 读取数据和简单的数据探索


```python
import numpy as np 
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 安装scikit learn
# conda install -c anaconda scikit-learn
# 主要使用datasets，里面就是数据集
from sklearn import datasets
```

```python
# iris就是鸢尾花的数据集
# 可以理解为一个字典
iris = datasets.load_iris()
print(iris.data)
```

```python
# data: 数据
# feature_names：样本特征名称
# target: y
# target_names: y的名称
# DESCR: 描述文档
iris.keys() 
```

```python
print(iris.DESCR)
```

```python
iris.target_names
```

```python
X = iris.data[:,:2]
```

```python
# 因为点之间连线没有意义，所以使用散点图
# 因为只能二维，所以取前两列
plt.scatter(X[:,0], X[:,1], label=iris.feature_names[1])
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.axis([4, 8, 1, 4.5])
plt.legend()
plt.show()
```

```python
y = iris.target
```

```python
# 种类为0的鸢尾花的0特征和1特征的散点图
plt.scatter(X[y==0,0], X[y==0,1], color="red", label=iris.target_names[0], marker="o")
# 种类为1的鸢尾花的0特征和1特征的散点图
plt.scatter(X[y==1,0], X[y==1,1], color="blue", label=iris.target_names[1], marker="+")
# 种类为2的鸢尾花的0特征和1特征的散点图
plt.scatter(X[y==2,0], X[y==2,1], color="green", label=iris.target_names[2], marker="x")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Different types of iris")
plt.legend()
plt.show()
```

```python
plt.scatter(X[y==0,0], X[y==0,1], color="red", marker="o")
plt.scatter(X[y==1,0], X[y==1,1], color="blue", marker="+")
plt.scatter(X[y==2,0], X[y==2,1], color="green", marker="x")
plt.show()
```

### 关于marker参数：[http://matplotlib.org/1.4.2/api/markers_api.html](http://matplotlib.org/1.4.2/api/markers_api.html)

```python
# 从索引2开始的所有数据
X = iris.data[:,2:]
```

```python
plt.scatter(X[y==0,0], X[y==0,1], color="red", marker="o")
plt.scatter(X[y==1,0], X[y==1,1], color="blue", marker="+")
plt.scatter(X[y==2,0], X[y==2,1], color="green", marker="x")
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.title("Different types of iris")
plt.show()
```
