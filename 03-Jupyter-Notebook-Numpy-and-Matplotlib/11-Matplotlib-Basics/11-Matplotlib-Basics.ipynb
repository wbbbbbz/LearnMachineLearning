## 11 ``matplotlib`` 基础

```python
import numpy as np 
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
# 其实使用mpl的机会不是非常多。mpl的功能非常强大
# 一般绘图使用子模块pyplot即可
import matplotlib as mpl
import matplotlib.pyplot as plt

# 第一个是横坐标，第二个是纵坐标
plt.plot(x, y)
```

```python
# show就是表达图像
plt.show()
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
```

```python
# 绘制多条曲线
plt.plot(x, siny)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red")
plt.show()
```

### 关于color参数：[https://matplotlib.org/2.0.2/api/colors_api.html](https://matplotlib.org/2.0.2/api/colors_api.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.show()
```

### 关于linestyle参数：[https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html](https://matplotlib.org/devdocs/gallery/lines_bars_and_markers/line_styles_reference.html)

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlim(-5, 15)
plt.ylim(0, 1)
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
# 直接对x,y两个轴进行调整
plt.axis([-1, 11, -2, 2])
plt.show()
```

```python
plt.plot(x, siny)
plt.plot(x, cosy, color="red", linestyle="--")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
plt.plot(x, siny, label="sin(x)")
plt.plot(x, cosy, color="red", linestyle="--", label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcome to matplotlib world!")
plt.show()
```

## Scatter Plot(散点图)

```python
plt.scatter(x, siny)
plt.show()
```

```python
plt.scatter(x, siny)
plt.scatter(x, cosy, color="red")
plt.show()
```

```python
x = np.random.normal(0, 1, 10000)
y = np.random.normal(0, 1, 10000)

# alpha是不透明度
plt.scatter(x, y, alpha=0.1)
plt.show()
```
