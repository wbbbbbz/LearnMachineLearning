## 什么是多项式回归

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-3, 3, size=100)
X = x.reshape(-1, 1)
# 二次函数生成y
y = 0.5 * x**2 + x + 2 + np.random.normal(0, 1, 100)
```

```python
plt.scatter(x, y)
plt.show()
```

### 线性回归？

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_predict = lin_reg.predict(X)
```

```python
plt.scatter(x, y)
plt.plot(x, y_predict, color='r')
plt.show()
```

```python
lin_reg.score(X, y)
```

### 解决方案， 添加一个特征

```python
X2 = np.hstack([X, X**2])
```

```python
# 相当于根据原有数据集创建新的数据集
# 每一个样本中有两个特征
X2.shape
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(X2, y)
y_predict2 = lin_reg2.predict(X2)
```

```python
plt.scatter(x, y)
# 因为x是没有顺序的，如果生成折线图的话需要进行排序
# 对应的y也要进行排列
plt.plot(np.sort(x), y_predict2[np.argsort(x)], color='r')
plt.show()
```

```python
lin_reg2.coef_
```

```python
lin_reg2.intercept_
```

```python
lin_reg2.score(X2, y)
```
