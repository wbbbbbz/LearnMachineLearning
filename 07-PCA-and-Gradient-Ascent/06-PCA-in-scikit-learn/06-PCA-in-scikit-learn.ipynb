## scikit-learn中的PCA

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=666)
```

```python
X_train.shape
```

```python
%%time

from sklearn.neighbors import KNeighborsClassifier

knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train, y_train)
```

```python
knn_clf.score(X_test, y_test)
```

```python
from sklearn.decomposition import PCA

# 从64维降到2维
pca = PCA(n_components=2)
pca.fit(X_train)
X_train_reduction = pca.transform(X_train)
# 测试数据集也要使用同样的pca
X_test_reduction = pca.transform(X_test)
```

```python
# 训练速度是变快了
%%time 
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_reduction, y_train)
```

```python
# 识别精度下降的太低了
knn_clf.score(X_test_reduction, y_test)
```

### 主成分所解释的方差

```python
# 解释的方差比例
# 现在的pca有两个维度，第一维度大概解释14.5%，第二维度大概解释了13.7%
# 大概解释了28%左右，这样的话丢失的方差过多
pca.explained_variance_ratio_
```

```python
pca.explained_variance_
```

```python
from sklearn.decomposition import PCA

# 直接pca64个轴
pca = PCA(n_components=X_train.shape[1])
pca.fit(X_train)
# 直接pca所有的特征，就可以发现每一个轴对应的重要程度
pca.explained_variance_ratio_
```

```python
# 前i个轴的解释力度之和
plt.plot([i for i in range(X_train.shape[1])], 
         [np.sum(pca.explained_variance_ratio_[:i+1]) for i in range(X_train.shape[1])])
plt.show()
```

```python
# scikit-learn中直接封装了该功能
# 输入小数。虽然不知道需要多少维，但是需要解释95%的方差
pca = PCA(0.95)
pca.fit(X_train)
```

```python
# 使用28维的数据即可
pca.n_components_
```

```python
X_train_reduction = pca.transform(X_train)
X_test_reduction = pca.transform(X_test)
```

```python
%%time 
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train_reduction, y_train)
```

```python
# 精度稍微的丢失，但是换来时间大幅度缩短
knn_clf.score(X_test_reduction, y_test)
```

### 使用PCA对数据进行降维可视化
压缩到2维不是没有意义

```python
pca = PCA(n_components=2)
pca.fit(X)
X_reduction = pca.transform(X)
```

```python
for i in range(10):
    plt.scatter(X_reduction[y==i,0], X_reduction[y==i,1], alpha=0.8)
plt.show()
# 每一个颜色就有不同的颜色。在降到二维的时候不同数字已经有很大的区分度了
# 通过二维的可视化，已经能得到一些提示。比如某两个数字之间如果需要分类的话2维已经足够了
```
